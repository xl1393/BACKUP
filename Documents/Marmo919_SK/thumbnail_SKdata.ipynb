```python
from os import listdir
from os.path import isfile, join
import h5py
import cv2
import numpy as np
from skimage.util import view_as_blocks
from time import gmtime, strftime
from multiprocessing import Pool
from os import mkdir
import glob, os
```

```python
def FileProcess(i):
    print 'processing number '+str(i)
    #folder='/home/xli/M919_Green_SK/'+str(i)+'/'
    folder = '/scratch/PMD1463_SK/'+str(i)+'/'
    k=np.zeros((256,256,300))
    os.chdir(folder)
    kcount=0
    for file in glob.glob("*.jp2"):
        img=cv2.imread(folder+file,0)
        temp = np.zeros([256,256])
        temp[:img.shape[0],:img.shape[1]] = img
        
        k[:,:,kcount]=temp
        kcount=kcount+1
        
        
        thumbnail=np.max(k,axis=2)
        cv2.imwrite('/scratch/PMD_1463_Thumbnail/'+str(i)+'.jpg',thumbnail)
```

```python
data=range(1831)
```

```python
p=Pool(20)
p.map(FileProcess,data)
```
